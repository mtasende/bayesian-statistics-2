```python
import re

month_matcher = re.compile('\d{4}M\d{2}')

months = [month_matcher.match(col).group(0) for col in cpi_usa_df.columns if month_matcher.match(col) is not None]

cpi_usa_df = cpi_usa_df.loc[:, months]

def transform_month(weird_str):
    year, month = weird_str.split('M')
    return dt.date(int(year), int(month), 15)

cpi_usa_df.rename(columns={old:transform_month(old) for old in cpi_usa_df.columns}, inplace=True)
```
