```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import sys
from time import time
import re
import datetime as dt

%matplotlib inline

%pylab inline
pylab.rcParams['figure.figsize'] = (20.0, 10.0)

%load_ext autoreload
%autoreload 2

# Constants definition
DATA_PATH = '../data/preprocessed_data/'
PROJECT_ROOT = '../'
SRC = PROJECT_ROOT + 'src/'

RATES = 'rates.csv'
CODES = 'codes.csv'
CPI = 'cpi.csv'
CPI_USA = 'cpi_usa.csv'

sys.path.append(SRC)
```

```python
rates_df = pd.read_csv(DATA_PATH + RATES, index_col=0)
codes_df = pd.read_csv(DATA_PATH + CODES, index_col=0)
cpi_df = pd.read_csv(DATA_PATH + CPI, index_col=0)
cpi_usa_df = pd.read_csv(DATA_PATH + CPI_USA, index_col=0)
```

## Let's try to join all that in one dataframe

```python
s_rates_df = pd.DataFrame(rates_df.stack()).rename(columns={0:'exchange_rate'})
s_rates_df['symbol'] = s_rates_df.index.get_level_values(1)
s_rates_df['date'] = s_rates_df.index.get_level_values(0)
print(s_rates_df.shape)
s_rates_df.head()
```

```python
s_rates_df.index.get_level_values(1)
```

```python
s_cpi_df = pd.DataFrame(cpi_df.stack()).rename(columns={0:'cpi'})
print(s_cpi_df.shape)
s_cpi_df.head()
```

```python
data_df = s_rates_df.join(s_cpi_df)
data_df.index = data_df.index.droplevel(0)
data_df
```

```python
data_df = data_df.join(codes_df).reset_index().drop('index', 1)
data_df.head()
```

## Let's add the USA cpi

```python
cpi_usa_df = cpi_usa_df.transpose().rename(columns={'USD':'cpi_usa'})
cpi_usa_df.head()
```

```python
data_df = data_df.groupby('symbol').apply(lambda x: x.set_index('date', drop=False).join(cpi_usa_df))
data_df
```

```python
# data_df.reset_index(level=1, inplace=True, drop=True)
data_df.reset_index(inplace=True, drop=True)
data_df.head()
```
